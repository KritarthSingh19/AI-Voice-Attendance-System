```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

file_path = "../data/enrollment/kritarth_singh/sample_01.wav"

y, sr = librosa.load(file_path, sr=16000)

print("Sample Rate:", sr)
print("Audio Length (samples):", len(y))
print("Duration (sec):", len(y)/sr)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(y)
plt.title("Waveform of Speech Signal")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.show()
```

```python
D = librosa.amplitude_to_db(
    np.abs(librosa.stft(y)),
    ref=np.max
)

plt.figure(figsize=(12, 4))
plt.imshow(D, aspect='auto', origin='lower')
plt.title("Spectrogram (Frequency vs Time)")
plt.colorbar()
plt.show()
```

```python
mfcc = librosa.feature.mfcc(
    y=y,
    sr=sr,
    n_mfcc=13
)

print("MFCC Shape:", mfcc.shape)

plt.figure(figsize=(12,4))
plt.imshow(mfcc, aspect='auto', origin='lower')
plt.title("MFCC Features")
plt.colorbar()
plt.show()
```

```python
mfcc_mean = np.mean(mfcc, axis=1)

print("Feature Vector Shape:", mfcc_mean.shape)
print(mfcc_mean)
```
